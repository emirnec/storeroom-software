```python
from tkinter import filedialog
import cv2
import tkinter as tk
from tkinter import simpledialog
from abc import ABC
from abc import abstractmethod
import pandas as pd
from collections.abc import MutableMapping
import xlwt
from tkinter import Radiobutton
import os
print(os.getcwd()) #we used this code to find where our script is working from 



class Visual:
    def __init__(self, master):
        self.master = master
        self.camera_button = tk.Button(self.master, text="Open Camera", command=self.open_camera)
        self.camera_button.pack()
        self.image_button = tk.Button(self.master, text="Open Image", command=self.open_image)
        self.image_button.pack()
        self.cap = None
        self.password = "4820"

    def open_camera(self):
        entered_password = simpledialog.askstring("Password", "Enter the password:", show='*')
        if entered_password != self.password:
            print("Incorrect password.")
            return
        self.cap = cv2.VideoCapture(0)
        if not self.cap.isOpened():
            print("Cannot open camera")
            return
        self.camera_button.config(state='disable')
        while True:
            ret, frame = self.cap.read()
            if not ret:
                break
            cv2.imshow("Camera", frame)
            if cv2.waitKey(1) == ord('q'):
                break
        self.cap.release()
        cv2.destroyAllWindows()
        self.camera_button.config(state='normal')

    def open_image(self):
        try:
            image = cv2.imread("Taylor's Fabrics.jpg ",1)
            cv2.imshow("Taylor's Tailor", image)
            cv2.waitKey(0)
        except:
            print("Error: Could not read or find the image.")
        finally:
            cv2.destroyAllWindows()


class RawMaterials:
    def __init__(self):
        self.store = dict()

    def __getitem__(self, key):
        return self.store[key]

    def __setitem__(self, key, value):
        self.store[key] = value

    def __delitem__(self, key):
        del self.store[key]

    def __iter__(self):
        return iter(self.store)

    def __len__(self):
        return len(self.store)
    
    def set_raw_material(self, name, date_of_purchase, supplier_name, storage_expiration, storage_code, description):
        self.store[storage_code] = {"name": name,
                                   "date_of_purchase": date_of_purchase,
                                   "supplier_name": supplier_name,
                                   "storage_expiration": storage_expiration,                              
                                   "description": description}
   
    def get_raw_material(self, storage_code):
        return self.store.get(storage_code)
       
    def get_raw_materials(self):
        return self.store.values()


class Products(RawMaterials):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

    def __keytransform__(self, key):
        return key.lower()

    def save_to_excel(self, file_name):
        workbook = xlwt.Workbook()
        worksheet = workbook.add_sheet("Products")
        row = 0
        for key, value in self.store.items():
            col = 0
            for val in value.values():
                worksheet.write(row, col, val)
                col += 1
            row += 1
        workbook.save(file_name)

    def load_from_excel(self, file_name):
        pass
       
    def get_raw_materials(self):
        return self.get_raw_materials
       
    def set_product(self, name, date_of_production, customer_name, product_expiration, storage_code, raw_material_codes, description):
        self.get_products.append(
            {"name": name,
            "date_of_production": date_of_production,
            "customer_name": customer_name,
            "product_expiration": product_expiration,
            "storage_code": storage_code,
            "raw_material_codes": raw_material_codes,
            "description": description})
   
    def get_product(self, storage_code):
        for product in self.get_products:
            if product["storage_code"] == storage_code:
                return product
   
    def get_products(self):
        return self.get_products
    



store = RawMaterials()

store.set_raw_material("Cotton", "01.01.2021", "Descrip DS", "11/04/2024", "Fabric", "White cotton fabric")
store.set_raw_material("Denim", "01.01.2021", "Descrip DS", "11/04/2024", "Fabric", "Black and nlue denim fabric")
store.set_raw_material("Leather", "01.01.2021", "Descrip DS", "11/04/2024", "Fabric", "%100 Animal leather")
store.set_raw_material("Satin", "01.01.2021", "Descrip DS", "11/04/2024", "Fabric", "")
store.set_raw_material("Lace", "01.01.2021", "Descrip DS", "11/04/2024", "Fabric", "")


products = Products()

products.set_raw_material("Jackket", "02.01.2023", "İsa Özgün", "13.12.2023", "EN06", "%80 Satin and %20 Cotton Jacket")  
products.set_raw_material("Gloves", "14.05.2023", "Mahmut Arikboğa", "12.01.2023", "MA21", "%100 Leather Gloves")
products.set_raw_material("Jean", "02.01.2022", "Enes Taş", "27.04.2022", "IA63", "%100 Denim Jean")
products.set_raw_material("Sweatshirt", "30.03.2023", "Mehmet Sarp", "19.08.2023", "ES48", "%100 Cotton Sweatshirt")
products.set_raw_material("Jogger", "06.06.2021", "Baran Alphan", "22.03.2022", "ET34", "%80 Cotton and %20 Lace Jogger")

products.save_to_excel("Taylor's Tailor.xls")

master = tk.Tk()
canvas = tk.Canvas (master, height=500, width=800)
canvas.pack()
frame_back = tk.Frame(master, bg='#9400d3')
frame_back.place(relx=0, rely=0, relwidth=1, relheight=1)
frame_up = tk.Frame(master, bg='#191970')
frame_up.place(relx=0.005, rely=0.005, relwidth=0.99, relheight=0.2)
frame_left = tk.Frame(master, bg='#191970')
frame_left.place(relx=0.005, rely=0.21, relwidth=0.23, relheight=0.785)
frame_right = tk.Frame(master, bg='#191970')
frame_right.place(relx=0.239, rely=0.21, relwidth=0.756, relheight=0.785)

var=tk.IntVar()

R1 = Radiobutton(frame_left, text="COTTON", variable = var, value=1, bg='#e0ffff', font="Times 11 italic")
R1.pack(pady=5,padx=15)

R2 = Radiobutton(frame_left, text="DENIM", variable = var, value=2, bg='#e0ffff', font="Times 11 italic")
R2.pack(pady=5,padx=15)

R3 = Radiobutton(frame_left, text="LEATHER", variable = var, value=3, bg='#e0ffff', font="Times 11 italic")
R3.pack(pady=5,padx=15)

R4 = Radiobutton(frame_left, text="SATIN", variable = var, value=4, bg='#e0ffff', font="Times 11 italic")
R4.pack(pady=5,padx=15)

R5 = Radiobutton(frame_left, text="LACE", variable = var, value=5, bg='#e0ffff', font="Times 11 italic")
R5.pack(pady=5,padx=15)

from tkinter import messagebox

def send():
    last_message=""

    if var.get():
        if var.get()==1:
            last_message+="01.01.2021 Descrip DS 11/04/2024 1111 Fabric"
        elif var.get()==2:
            last_message+="01.01.2021 Descrip DS 11/04/2024 1112 Fabric"
        elif var.get()==3:
            last_message+="01.01.2021 Descrip DS 11/04/2024 1113 Fabric"
        elif var.get()==4:
            last_message+="01.01.2021 Descrip DS 11/04/2024 1114 Fabric"
        elif var.get()==5:
            last_message+="01.01.2021 Descrip DS 11/04/2024 1115 Fabric"
        messagebox.showinfo("expiration time", last_message)
        
    return

check_button=tk.Button(frame_left, text="check", command=send)
check_button.pack(padx=5,pady=15)

visual = Visual(master) 
master.mainloop()



```
